```python
from __future__ import annotations

from typing import Callable, Dict, List, Tuple
import numpy as np
import pandas as pd
from scipy import stats

# ---------- helpers ----------
def _paired_clean(a: np.ndarray, b: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    m = np.isfinite(a) & np.isfinite(b)
    return a[m], b[m]

def _bootstrap_ci(
    stat_fn: Callable[[np.ndarray, np.ndarray], float],
    a: np.ndarray,
    b: np.ndarray,
    n_boot: int = 2000,
    seed: int = 1706,
) -> Tuple[float, float, float]:
    rng = np.random.default_rng(seed)
    a, b = _paired_clean(a, b)
    n = len(a)
    if n < 3:
        return np.nan, np.nan, np.nan

    vals = []
    for _ in range(n_boot):
        idx = rng.integers(0, n, size=n)
        v = stat_fn(a[idx], b[idx])
        if np.isfinite(v):
            vals.append(v)

    if not vals:
        return np.nan, np.nan, np.nan

    lo, hi = np.percentile(vals, [2.5, 97.5])
    return float(np.mean(vals)), float(lo), float(hi)

def _pearson(a: np.ndarray, b: np.ndarray) -> float:
    a, b = _paired_clean(a, b)
    if len(a) < 3 or np.allclose(a, a[0]) or np.allclose(b, b[0]):
        return np.nan
    return float(np.corrcoef(a, b)[0, 1])

def _spearman(a: np.ndarray, b: np.ndarray) -> float:
    a, b = _paired_clean(a, b)
    if len(a) < 3:
        return np.nan
    r, _ = stats.spearmanr(a, b, nan_policy="omit")
    return float(r)

def icc2_1(a: np.ndarray, b: np.ndarray) -> float:
    """
    ICC(2,1): two-way random effects, absolute agreement, single measurement.
    """
    a, b = _paired_clean(a, b)
    n = len(a)
    k = 2
    if n < 3:
        return np.nan

    Y = np.vstack([a, b]).T  # (n, 2)
    mean_subject = Y.mean(axis=1, keepdims=True)
    mean_rater = Y.mean(axis=0, keepdims=True)
    grand_mean = Y.mean()

    SSR = k * np.sum((mean_subject - grand_mean) ** 2)
    SSC = n * np.sum((mean_rater - grand_mean) ** 2)
    SSE = np.sum((Y - mean_subject - mean_rater + grand_mean) ** 2)

    MSR = SSR / (n - 1)
    MSC = SSC / (k - 1)
    MSE = SSE / ((n - 1) * (k - 1))

    denom = MSR + (k - 1) * MSE + (k * (MSC - MSE) / n)
    if denom <= 0:
        return np.nan
    return float((MSR - MSE) / denom)

def _mean_shift(a: np.ndarray, b: np.ndarray) -> float:
    a, b = _paired_clean(a, b)
    if len(a) == 0:
        return np.nan
    return float(np.mean(b - a))  # B - A

def _median_min_max(x: np.ndarray) -> Tuple[float, float, float]:
    x = np.asarray(x, dtype=float)
    x = x[np.isfinite(x)]
    if len(x) == 0:
        return np.nan, np.nan, np.nan
    return float(np.median(x)), float(np.min(x)), float(np.max(x))

# ---------- main ----------
def check_statistic_two_dfs(
    df_a: pd.DataFrame,
    df_b: pd.DataFrame,
    *,
    features: List[str] = ["neg", "neu", "pos", "compound"],
    align: str = "inner",          # "inner" (intersection of indices) / "left" / "right"
    agg: str | None = None,        # None -> compare row by row; "mean"/"median" -> first aggregate each DF into one row
    n_boot: int = 2000,
    seed: int = 1706,
) -> pd.DataFrame:
    # 1) keep only the required columns
    a = df_a[[c for c in features if c in df_a.columns]].copy()
    b = df_b[[c for c in features if c in df_b.columns]].copy()

    # 2) convert to numeric
    for c in a.columns:
        a[c] = pd.to_numeric(a[c], errors="coerce")
    for c in b.columns:
        b[c] = pd.to_numeric(b[c], errors="coerce")

    # 3) aggregate (if needed)
    if agg is not None:
        if agg not in ("mean", "median"):
            raise ValueError("agg must be one of: None, 'mean', 'median'")
        a = getattr(a, agg)(axis=0, numeric_only=True).to_frame().T
        b = getattr(b, agg)(axis=0, numeric_only=True).to_frame().T

    # 4) align by index (for row-wise comparison)
    if agg is None:
        a, b = a.align(b, join=align, axis=0)

    rows: List[Dict[str, object]] = []
    for feat in features:
        if feat not in a.columns or feat not in b.columns:
            rows.append({"feature": feat, "n": 0, "note": "missing column"})
            continue

        x = a[feat].to_numpy(dtype=float)  # A
        y = b[feat].to_numpy(dtype=float)  # B
        x, y = _paired_clean(x, y)
        n = len(x)

        a_med, a_min, a_max = _median_min_max(x)
        b_med, b_min, b_max = _median_min_max(y)

        r_m, r_lo, r_hi = _bootstrap_ci(_pearson, x, y, n_boot=n_boot, seed=seed)
        s_m, s_lo, s_hi = _bootstrap_ci(_spearman, x, y, n_boot=n_boot, seed=seed)
        icc_m, icc_lo, icc_hi = _bootstrap_ci(lambda p, q: icc2_1(p, q), x, y, n_boot=n_boot, seed=seed)
        sh_m, sh_lo, sh_hi = _bootstrap_ci(_mean_shift, x, y, n_boot=n_boot, seed=seed)

        rows.append({
            "feature": feat,
            "n": int(n),
            "A_median[min,max]": f"{a_med:.3f} [{a_min:.3f},{a_max:.3f}]",
            "B_median[min,max]": f"{b_med:.3f} [{b_min:.3f},{b_max:.3f}]",
            "r [95% CI]": f"{r_m:.3f} [{r_lo:.3f},{r_hi:.3f}]",
            "rho [95% CI]": f"{s_m:.3f} [{s_lo:.3f},{s_hi:.3f}]",
            "ICC(2,1) [95% CI]": f"{icc_m:.3f} [{icc_lo:.3f},{icc_hi:.3f}]",
            "mean shift Δμ (B-A) [95% CI]": f"{sh_m:.3f} [{sh_lo:.3f},{sh_hi:.3f}]",
            "note": "",
        })

    return pd.DataFrame(rows)
```
